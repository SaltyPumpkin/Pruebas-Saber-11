```python
# Comparaci√≥n de medias entre Montenegro y Calarc√° usando prueba t
t_stat, p_value = ttest_ind(
    Quindio_Montenegro_resultados['PUNT_GLOBAL'],
    Quindio_Calarca_resultados['PUNT_GLOBAL'],
    equal_var=True  # Usa False si las varianzas son diferentes
)

# Resultados
alpha = 0.05  # Nivel de significancia
print(f'Estad√≠stico t: {t_stat}')
print(f'Valor p: {p_value}')
if p_value < alpha:
    print("Rechazamos la hip√≥tesis nula: las medias son significativamente diferentes.")
else:
    print("No rechazamos la hip√≥tesis nula: no hay evidencia suficiente para decir que las medias son diferentes.")
```

```python
Gen_Quindio = Datos[['ESTU_GENERO', 'ESTU_COD_RESIDE_DEPTO', 'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES', 'PUNT_GLOBAL']]
Gen_Quindio = Gen_Quindio.loc[(Gen_Quindio['ESTU_COD_RESIDE_DEPTO'] == '63') | (Gen_Quindio['ESTU_COD_RESIDE_DEPTO'] == 63)]

Gen_F_Quindio = Gen_Quindio.loc[(Gen_Quindio['ESTU_GENERO'] == 'F')]
Gen_M_Quindio = Gen_Quindio.loc[(Gen_Quindio['ESTU_GENERO'] == 'M')]

X_Female_Quindio_mean = Gen_F_Quindio['PUNT_GLOBAL'].mean()
n_Female_Quindio_count = Gen_F_Quindio['PUNT_GLOBAL'].count()

# Comparaci√≥n de medias entre Montenegro y Calarc√° usando prueba t
t_stat_2, p_value_2 = ttest_ind(
    Gen_F_Quindio['PUNT_GLOBAL'],
    Gen_M_Quindio['PUNT_GLOBAL'],
    equal_var=True  # Usa False si las varianzas son diferentes
)

# Resultados
alpha = 0.05  # Nivel de significancia
print(f'Estad√≠stico t: {t_stat_2}')
print(f'Valor p: {p_value_2}')
if p_value < alpha:
    print("Rechazamos la hip√≥tesis nula: las medias son significativamente diferentes.")
else:
    print("No rechazamos la hip√≥tesis nula: no hay evidencia suficiente para decir que las medias son diferentes.")
```
